```python
year_2022_23 = df[df['Year'] == '2022-23']

categories = ['Readymade Garment', 'Cotton Textiles', 'Man-made Textiles', 'Wool & Woolen Textiles',
              'Silk Products', 'Handloom Products', 'Carpets', 'Jute Products']
sizes = [year_2022_23[category].sum() for category in categories]
labels = [category.replace('&', '&\n') for category in categories]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Textile & Apparel Production in 2022-23')
plt.tight_layout()
plt.show()
```

```python
categories = ['Readymade Garment', 'Cotton Textiles', 'Man-made Textiles', 'Wool & Woolen Textiles',
              'Silk Products', 'Handloom Products', 'Carpets', 'Jute Products']
totals = [df[category].sum() for category in categories]
labels = [category.replace('&', '&\n') for category in categories]  # Adjusting labels for better readability

# Creating the pie chart
plt.figure(figsize=(10, 8))
plt.pie(totals, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Total Textile & Apparel Production Across All Years')
plt.show()
```
